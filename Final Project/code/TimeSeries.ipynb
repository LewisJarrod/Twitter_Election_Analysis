```python
import pandas as pd 
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

df = pd.read_csv('data_reservoir_sample/times_data.csv')
```

```python
df_t = df.T
df_t = df_t.sort_index()
plt.figure(figsize=(200,100)); df_t.plot();
```

```python
df_t
# df_t.reset_index(level=0, inplace=True)
# df_t = df_t.drop('index', 1 )
df_t
data = []
for index, row in df_t.iterrows():
    time = (str(index)[:-2])
    if int(time[:-2]) < 30:
        time = time[:-2] + '00'
    else :
        time = time[:-2] + '30'
    str_time = time[:2] + '/' + time[2:4] + '/' + time[4:]

    data.append((time,str_time, row[0]))
new_df = pd.DataFrame(data, columns =['time','str_time', 'count'])
new_df 
```

```python
total_df = new_df.groupby(['time'],as_index=False).sum()
x = total_df['time']
y = total_df['count']
plt.plot(x, y, color='#EC407A', linewidth=2.0)


# for name, group in groups:
#     if name not in [0, 1, 2]: 
#         ax.plot(group.x_vals, group.y_vals, marker='o', linestyle='', ms=8, label=cluster_names[name], \
#             color=colors[name], mec='none', alpha=.08)
#     else:
#         ax.plot(group.x_vals, group.y_vals, marker='o', linestyle='', ms=10, label=cluster_names[name], \
#                 color=colors[name], mec='none', alpha=.5)
#         ax.set_aspect('auto')
#         ax.tick_params(axis= 'x', which='both', bottom='off', top='off', labelbottom='off')
#         ax.tick_params(axis= 'y', which='both', left='off', top='off', labelleft='off')
    
# legend = ax.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=2, borderaxespad=0,prop={'size':20})
# ax.set_axis_bgcolor('white')

# plt.show()

plt.show()
```
